```python
import sys

sys.path.append("../src")

from pathlib import Path
from pprint import pprint
from collections import defaultdict
from tqdm import tqdm
import pickle
from copy import deepcopy
```

```python
binary = pickle.loads(Path("../stats/binary_modify.pkl").read_bytes())
```

```python
def modify_prob(binary_prob: dict):
    binary_prob = deepcopy(binary_prob)
    modified = defaultdict(list)
    for _parent, children in tqdm(binary_prob.items()):
        for prob, (left, right) in children:
            parent = _parent
            if parent.endswith(">") or not right.endswith(">"):
                modified[parent].append((prob, prob, (left, right)))
                continue

            
            modified_prob = prob**(1 / (right.count("-") + 1))
            while True:
                modified[parent].append((prob, modified_prob, (left, right)))
                parent = right
                # print(parent, binary_prob[parent])
                _, (left, right) = binary_prob[parent][0]
                if not right.endswith(">"):
                    break
            modified[parent].append((prob, modified_prob, (left, right)))
    return modified
```

```python
modified = modify_prob(binary)
```
